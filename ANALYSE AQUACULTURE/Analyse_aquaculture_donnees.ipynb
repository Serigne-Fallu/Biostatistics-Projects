### ANALYSE GENERALE DE LA CORRELATION ENTRE TOUTES LES VAROABLES QUQNTITATIVES:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_excel('JEUX DE_DONNEES_AQUA_CORR.xlsx')
data_cont = data.drop(["Date", "Aliment", "Bacs"], axis=1)

correlation = data_cont.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Corrélation entre les variables')
plt.show()
```

### ANALYSE CORRELATION ENTRE VARIABLES ALIMENTAIRES ET VARIABLES DE CROISSANCE:

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation = data[['Ration (g)',"TCA", "Gain Pm absolu (g) ", 'Taux de survie (%)', "Taux de croissance spécifique (% en j)"]].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Corrélation entre les variables d\'alimentation et les variables de croissance')
plt.show()

sns.pairplot(data_cont, x_vars=['Gain Pm absolu (g) ', 'TCA', 'Taux de survie (%)', "Taux de croissance spécifique (% en j)"], 
             y_vars=['Ration (g)'],kind='reg', height=3)
plt.title('Corrélation entre les variables d\'alimentation et les variables de croissance')
plt.show()
```

### COMPARAISON DES RESULTATS POUR CHAQUE TYPE ALIMENT:

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Charger le jeu de données (remplacez 'nom_du_fichier.xlsx' par le nom de votre fichier)
data = pd.read_excel('JEUX DE_DONNEES_AQUA_CORR.xlsx')
data['Gain Pm absolu (g) '] = pd.to_numeric(data['Gain Pm absolu (g) '], errors='coerce')
data['TCA'] = pd.to_numeric(data['TCA'], errors='coerce')
data['Taux de survie (%)'] = pd.to_numeric(data['Taux de survie (%)'], errors='coerce')
data['Ration (g)'] = pd.to_numeric(data['Ration (g)'], errors='coerce')
data['Ration (g)'] = pd.to_numeric(data['Taux de croissance spécifique (% en j)'], errors='coerce')


mapping = {'Local 0': 0, 'Local 10': 10, 'Local 25': 25}

data['Aliment'] = data['Aliment'].replace(mapping)
scatterplot = sns.pairplot(data[['Gain Pm absolu (g) ', 'TCA', 'Taux de survie (%)', 'Aliment', "Taux de croissance spécifique (% en j)"]], hue='Aliment')
plt.show()
```

```python
# Créer un graphe de type "boxplot" pour chaque variable en fonction de l'aliment
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(data=data, x='Aliment', y='Gain Pm absolu (g) ')
plt.subplot(2, 2, 2)
sns.boxplot(data=data, x='Aliment', y='TCA')
plt.subplot(2, 2, 3)
sns.boxplot(data=data, x='Aliment', y='Taux de survie (%)')
plt.subplot(2, 2, 4)
sns.boxplot(data=data, x='Aliment', y='Taux de croissance spécifique (% en j)')

plt.tight_layout()
plt.show()
```

```python
variables_quantitatives = ["Ration (g)", 'TCA', 'Gain Pm absolu (g) ','Taux de survie (%)',"Taux de croissance spécifique (% en j)","Croissance journalière (g/j)"]

# Création des sous-graphiques
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 8))
axes = axes.flatten()

# Parcours des variables quantitatives
for i, variable in enumerate(variables_quantitatives):
    ax = axes[i]
    data_grouped = data.groupby('Aliment')[variable].mean().reset_index()
    labels = data_grouped['Aliment']
    values = data_grouped[variable]
    ax.pie(values, labels=labels, autopct='%1.1f%%')
    ax.set_title(variable)
plt.suptitle('Comparaison des resultats trois types d\'aliments testés', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```
